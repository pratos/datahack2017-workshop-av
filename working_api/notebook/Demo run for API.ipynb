```python
import json
import requests
import pandas as pd
```

```python
"""Setting the headers to send and accept json responses
"""
header = {'Content-Type': 'application/json', \
                  'Accept': 'application/json'}

"""Reading test batch
"""
df = pd.read_csv('../test_data/test.csv', encoding="utf-8-sig")
df = df.head()

"""Converting Pandas Dataframe to json
"""
data = df.to_json(orient='records')
```

```python
data
```

```python
"""POST <url>/predict
"""
resp = requests.post("http://34.231.240.130:5000/predict", \
                    data = json.dumps(data),\
                    headers= header)
```
